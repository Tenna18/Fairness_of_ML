## Age investigation

```python
sns.distplot(df['Ages'],kde=True, bins=30)
```

```python
under4s = []

for i in range(len(df)):
    if df['Ages'].iloc[i] <= 4:
        under4s.append(df.iloc[i])
under4s = pd.DataFrame(under4s)

under4s = under4s.sample(frac=0.3)

df_over4 = df[df['Ages'] > 4]

df_adj_age = pd.concat([df_over4, under4s], ignore_index = True)
```

```python
df_adj_age
```

```python
sns.distplot(df_adj_age['Ages'],kde=True, bins=30)
```
